```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("C:/Users/osman/Desktop/goit/python/applications.csv")
industries = pd.read_csv("C:/Users/osman/Desktop/goit/python/industries.csv")
```

```python
df.head()
```

```python
industries.head()
```

```python
df = df.drop_duplicates(subset="applicant_id")
```

```python
df["External Rating"] = df["External Rating"].fillna(0)
```

```python
df["Education level"] = df["Education level"].fillna("Ortaokul")
```

```python
df
```

```python
df.isnull().sum()
```

```python
df = df.merge(industries, on="Industry", how="left")
```

```python
df.head()
```

```python
print(df.dtypes)
```

```python
df['Applied at'] = pd.to_datetime(df['Applied at'], format='mixed')
```

```python
print(df.dtypes)
```

```python
age_points = ((df['Age'] >= 35) & (df['Age'] <= 55)) * 20
weekday_points = (df['Applied at'].dt.dayofweek < 5) * 20  # 0-4 arası hafta içidir
marital_points = (df['Marital status'] == 'Married') * 20
location_points = (df['Location'].str.contains('Київ|Київська', na=False, case=False)) * 10
rating_high_points = (df['External Rating'] >= 7) * 20
rating_low_points = (df['External Rating'] <= 2) * -20
```

```python
df
```

```python
df['Total Score'] = (
    age_points + 
    weekday_points + 
    marital_points + 
    location_points + 
    df['Score'] + 
    rating_high_points + 
    rating_low_points
)
df['Total Score'] = df['Total Score'].clip(0, 100)
df.loc[df['Amount'].isna() | (df['External Rating'] == 0), 'Total Score'] = 0
```

```python
accepted_apps = df[df['Total Score'] > 0].copy()
accepted_apps['Week'] = accepted_apps['Applied at'].dt.isocalendar().week
weekly_avg = accepted_apps.groupby('Week')['Total Score'].mean()
```

```python
plt.figure(figsize=(10, 6))
weekly_avg.plot(kind='line', marker='o', color='teal')
plt.title('Haftalık Kabul Edilen Başvuruların Ortalama Puanı')
plt.xlabel('Hafta Numarası')
plt.ylabel('Ortalama Puan')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
accepted_apps = df[df['Total Score'] > 0].copy()
accepted_apps['Year'] = accepted_apps['Applied at'].dt.isocalendar().year
accepted_apps['Week'] = accepted_apps['Applied at'].dt.isocalendar().week

weekly_avg = accepted_apps.groupby(['Year', 'Week'])['Total Score'].mean().reset_index()
weekly_avg = weekly_avg.sort_values(['Year', 'Week'])
weekly_avg['Label'] = weekly_avg['Week'].astype(str) + ". Hafta"

color_map = {2022: 'blue', 2023: 'green'}
bar_colors = [color_map[year] for year in weekly_avg['Year']]

plt.figure(figsize=(14, 7))
bars = plt.bar(weekly_avg['Label'], weekly_avg['Total Score'], 
               color=bar_colors, edgecolor='black', alpha=0.8)

for bar in bars:
    yval = bar.get_height()
    if yval > 0: 
        plt.text(bar.get_x() + bar.get_width()/2, yval - 2,
                 f'{yval:.1f}', 
                 ha='center', va='top', 
                 color='white', fontweight='bold', fontsize=11)

plt.title('Yıla Göre Renklendirilmiş Haftalık Ortalama Puanlar', fontsize=15, pad=20)
plt.xlabel('Haftalar', fontsize=12)
plt.ylabel('Ortalama Toplam Puan', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.3)

from matplotlib.lines import Line2D
legend_elements = [Line2D([0], [0], color='blue', lw=4, label='2022 Yılı'),
                   Line2D([0], [0], color='green', lw=4, label='2023 Yılı')]
plt.legend(handles=legend_elements, loc='upper left')

plt.tight_layout()
plt.show()
```
